```python
#Part 2: more-messy-data.csv
df = pd.read_csv('more-messy-data.csv')
df
```

```python
duplicate_rows = df[df.duplicated()]
print("Number of duplicate rows in DF:", len(duplicate_rows))
```

```python
df = df.drop_duplicates()
```

```python
sepal_length_mean = df['sepal_length'].mean()
sepal_width_mean = df['sepal_width'].mean()
petal_length_mean = df['petal_length'].mean()
petal_width_mean = df['petal_width'].mean()

df['sepal_length'].fillna(sepal_length_mean, inplace=True)
df['sepal_width'].fillna(sepal_width_mean, inplace=True)
df['petal_length'].fillna(petal_length_mean, inplace=True)
df['petal_width'].fillna(petal_width_mean, inplace=True)

df
```

```python
missing_values = df.isnull().sum()
print(missing_values)
df.dropna()
```

```python
number_of_different_species = df['species'].count()
print("Number of different species:", number_of_different_species)
```

```python
df['species'] = df['species'].replace('setosa', 'setosa_species')
df['species'] = df['species'].replace('virginica', 'virginica_species')
df['species'] = df['species'].replace('Versicolor', 'Versicolor_species')

df
```
