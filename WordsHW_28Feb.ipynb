```python
#Data type of words
print(type(words))
```

```python
#Converts "english-words" to a list called "wordlist"
wordlist = list(words)
```

```python
#Prints the total number of wrods in the wordlist
print(len(wordlist))
```

```python
#Generates the first ten words in wordlist
print(wordlist[0:11])
```

```python
#Defines a list called "alpha_list", puts the words in alphabetical order, and generates the first 10
alpha_list = list(words)
alpha_list.sort()
print(alpha_list[0:11])
```

```python
#Caluclates the number of words that start with a
count = 0
for word in wordlist:
  if word.startswith('a'):
    count += 1
    count
    print("There are ", count, "words in the list that start with a")
```

```python
#Calculates the number of words that have a q
count = 0
for word in wordlist:
  if 'q' in words:
    count += 1
    count
    print("There are", count, "words in the list that have a q")
```

```python
count = 0
for word in wordlist:
  if 'e' not in words:
    count += 1
    count
    print("There are", count, "words in the list that do not have an e")
```
